```python
#Simulation of the actual game (first described on pages 7-8 of rubric)


glasses = int(input('How many glasses? ')) #choose how many glasses are on the table

#set up array for amount of liquid in each glass (indexed from 1 up to number of glasses)
table = np.zeros((glasses, 2))
for i in range(len(table)):
    table[i][0] = i + 1
print(table)



gameOver = False
while gameOver == False:
    
    #Ali's turn
    leftToPour = 0.5 #Ali can pour 1/2 pint
    while (leftToPour > 0.000001) and gameOver == False: #her go ends when she has poured all liquid, ue small number != 0 due to occassional rounding errors with fractions
        print()
        print(str(leftToPour) + " pint left in Ali's cup")
        cup = int(input("Which glass? ")) #choose cup
        
        
        #the values below ensure we enter the while loop
        num = 1
        den = 1
        
        
        #amount chosen must be >= what is left in Ali's cup
        while leftToPour - (num / den) < 0: #can't pour more than what's left in Ali's glass
            print()
            print(str(leftToPour) + " pint left in Ali's cup")
            num = float(input("Numerator? "))
            den = float(input("Denominator? "))
        
        #pour chosen amount in cup, reduce amount of water in Ali's cup
        table[cup - 1][1] += num / den
        leftToPour -= num / den
        print(table)
        
        #if cup over fills, game is over (Ali wins)
        if table[cup - 1][1] > 1:
            gameOver = True
    
    
    #Beth's turn
    if gameOver == False:
        
        #place holder values below ensure we enter the while loop
        c1 = 5
        c2 = 10
        
        while (c2 - c1 != 1) and (c1 != 1 or c2 != glasses): #while cups non-adjacent (1st cup adjacent to last cup too)
            #Beth chooses 2 cups
            print()
            print("Pick 2 ADJACENT cups to empty")
            c1 = int(input("Empty cup ...? "))
            c2 = int(input("and cup ...? "))
        
        #once valid cups are selected, they're emptied
        table[c1 - 1][1] = 0
        table[c2 - 1][1] = 0
        print()
        print(table)
        
    
    
    
    
    
#loop only ends if Ali wins    
print("Ali wins, glass " + str(cup) + " overflowed.")
```

```python
#Extend

g = int(input("Number of glasses"))

#create indexed array representing glasses
water = np.zeros((g, 2))
for i in range(g):
    water[i][0] = i + 1




print()
print("We start by keeping the water levels equal until each has 1/4 pint (which we know to be the limit)")
print()
print(water)

x = 0 #comparison values (while loop ends when we reach a limit)
y = x + 1

while np.abs(x - y) > 0.000000001: #being this close ~means equal
    y = x
    total = 0
    for i in range(g):
        total += water[i][1] #find total amount of water in all glasses
    x = (total + 0.5) / g #this will be the amount of water in each glass after Ali distributes her 1/2 pint
    for i in range(g):
        water[i][1] = x #adjust array accordingly

    
    
    #choose 2 random glasses to empty
    index = int(random.choice(np.arange(g)))
    water[index][1] = 0
    index = int(random.choice(np.setdiff1d(np.arange(g), [index])))
    water[index][1] = 0
    
        
        
    print()
    print(water)

    

    
print()
print()
print()
print("We have now reached 1/4 pint. Everytime Beth empties 2, we ignore them, and split the water between the glasses that still have water in them.")
print()


indexArr = np.zeros(g)
for i in range(len(indexArr)):    #list of non-empty glasses (will be changed every time Beth empties more)                         
    indexArr[i] = water[i][0]  #it is from this list that the 2 emptied glasses will randomly be selected


#We have reached our limit for amount of water in each glass when split equally
#Now we move onto the next part of the strategy - causing the overflow before Beth can empty every glass
while len(indexArr) != 1 and len(indexArr) != 2: #if we reach the point where there are 1 or 2 glasses and Ali doesn't have a glass > 1/2 pint, she has lost
    total = 0
    
    for i in range(g):
        total += water[i][1] #find total amount of water in all glasses
        
    x = (total + 0.5) / len(indexArr) #this will be the amount of water in each NON-EMPTY glass after Ali distributes her 1/2 pint (she is now ignoring empty glasses)
    for i in indexArr:
        water[int(i-1)][1] = x #adjust array accordingly

    #empty 2 random glasses
    #adjust list of non-empty glasses left
    for i in range(2):    
        index = int(random.choice(indexArr))
        water[index - 1][1] = 0
        indexArr = np.setdiff1d(indexArr, [index])
    print()
    print(water)

print()
print("It is now Ali's turn; if there is atleast 1 glass with over 1/2 pint then this strategy works for this number of glasses and Ali she has won")
```
